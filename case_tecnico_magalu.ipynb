### Análise de Custos Operacionais (`Base Case .xlsx`)

Carregamos a base de dados de custos operacionais e agrupamos os dados por mês para analisar a evolução dos custos de entrega, produtos e marketing. Os gráficos de linha mostram a tendência de crescimento para cada um desses custos ao longo do período.

```
import pandas as pd

df = pd.read_excel('Base Case .xlsx')
df.head()
```

```
import matplotlib.pyplot as plt

custo_mes = df.groupby('Mês')['Custo Entrega'].sum()
custo_mes.plot()

custo_prod = df.groupby('Mês')['Custo dos Produtos'].sum()
custo_prod.plot()

custo_mark = df.groupby('Mês')['Custo Marketing'].sum()
custo_mark.plot()
```

### Análise de Pedidos por Tipo de Entrega (`Base Case 2.xlsx`)

Em seguida, carregamos uma segunda base de dados para analisar o volume de pedidos. Agrupamos os dados para visualizar a distribuição de pedidos por tipo de entrega ao longo dos meses, comparando a performance e a popularidade de cada modalidade logística.

```
df = pd.read_excel('Base Case 2.xlsx')

df.head()
```

```
tipo_ped = df.groupby(['Mês','Tipo Entrega'])['Pedidos'].sum().unstack(level=0)

tipo_ped.plot()
```

Com base na análise, as conclusões principais são:

* **Crescimento de Custos e Pedidos:** A empresa está em expansão, com um aumento claro no volume de pedidos ao longo dos nove meses. Esse crescimento é acompanhado por um aumento nos custos de entrega e produtos, enquanto o investimento em marketing diminui.

* **"Retira Loja" Domina a Logística:** A modalidade "Retira Loja" é a preferida pela vasta maioria dos clientes, superando significativamente todos os outros métodos de entrega em todos os meses analisados.

* **Foco Estratégico:** A estratégia da empresa parece ser sustentar o crescimento de vendas com uma operação logística focada na retirada em loja, ao mesmo tempo em que otimiza ou reduz os custos de marketing.
